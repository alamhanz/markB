# Normalize

```python
import pandas as pd
import re
```

```python
PATH_RAW = '../data/raw/'
PATH_INTERIM = '../data/interim/'
```

```python
df_basket = pd.read_csv(PATH_INTERIM+'all_basket.csv')
```

## Check

```python
all_items = list(set(','.join(df_basket['0'].tolist()).split(',')))
all_items.sort()
```

```python
len(all_items)
```

```python
all_items
```

## Normalize

* Alpha numeric only
* lower strings

```python
# re.sub(r'\W+', ' ', 'asda/asfd (asdf)')
```

```python
all_items_norm = [re.sub(r'\W+', ' ', x.strip().lower()) for x in all_items]
```

```python
norm_rules = dict(zip(all_items,all_items_norm))
```

```python
def norm_items(basket):
    
    basket2 = [norm_rules[i] for i in basket.split(',')]
    basket2.sort()
    return ','.join(basket2)
```

```python
df_basket['norms'] = df_basket['0'].apply(norm_items)
```

```python
df_basket.to_csv(PATH_INTERIM+'all_basket2.csv', index = False)
```

## Split Data

```python
len(df_basket)*0.2
```

```python
all_unique_combination = df_basket.groupby('norms').size().sort_values()
all_unique_combination = all_unique_combination[all_unique_combination==1].sample(3500, random_state=532).index.tolist()
df_basket_test = df_basket[df_basket['norms'].isin(all_unique_combination)][['norms']]
df_basket_train = df_basket[~df_basket['norms'].isin(all_unique_combination)][['norms']]
```

```python
len(df_basket_train)+len(df_basket_test) == len(df_basket)
```

```python
df_basket_test.to_csv(PATH_INTERIM+'all_basket2_test.csv',index = False)
df_basket_train.to_csv(PATH_INTERIM+'all_basket2_train.csv',index = False)
```
